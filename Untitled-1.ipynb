```python
#include <stdio.h>

// Function declaration
int add(int x, int y);

int main() {
    int num1, num2, result;

    // Taking input from the user
    printf("Enter first number: ");
    scanf("%d", &num1);

    printf("Enter second number: ");
    scanf("%d", &num2);

    // Calling the function
    result = add(num1, num2);

    // Displaying the result
    printf("Sum = %d\n", result);

    return 0;
}

// Function definition
int add(int x, int y) {
    return x + y;
}
```
